```python
import numpy as np
import matplotlib.pyplot as plt
import random
%matplotlib inline
```

```python
x=np.linspace(0,10,10)
y1=x
y2=x**2
y3=x**3
y4=np.sqrt(x)

plt.figure()

plt.subplot(2,2,1)
plt.plot(x,y1,'ro' ) #marker='o', linestyle=':'
plt.title('$y1=x$')
plt.xlabel('X')
plt.ylabel('Y1')


plt.subplot(2,2,2)
plt.plot(x,y2,'g--' ) #marker='o', linestyle=':'
plt.title('$y2=x^2$')
plt.xlabel('X')
plt.ylabel('Y2')

plt.subplot(2,2,3)
plt.plot(x,y3,'b^' ) #marker='o', linestyle=':'
plt.title('$y_3=x^3$')
plt.xlabel('X')
plt.ylabel('Y3')

plt.subplot(2,2,4)
plt.plot(x,y4,'ks' ) #marker='o', linestyle=':'
plt.title('$y_4=x^1/2$')
plt.xlabel('X')
plt.ylabel('Y4')

# plt.legend()
plt.tight_layout()
# plt.show()
```
