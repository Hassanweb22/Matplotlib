```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[344,566,766,866,896]
z=[272,288,384,486,580]

plt.title('Yotube Views')
plt.xlabel('X')
plt.ylabel('Views')
plt.plot(x,y, label='Video 1')
plt.plot(x,z, label='Video 2')
plt.legend(loc='best', shadow=True)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[344,566,766,866,896]
z=[272,288,384,486,580]
f=[372,488,584,686,780]

plt.title('Yotube Views')
plt.xlabel('X')
plt.ylabel('Views')
plt.plot(x,y, label='Video 1', marker='o', markerfacecolor='blue',  markersize=6)
plt.plot(x,z, label='Video 2', marker='o', markerfacecolor='orange', markersize=6 )
plt.plot(x,f, label='Video 3', marker='o', markerfacecolor='green' , markersize=6)
# bbox_to_anchor=[0.5,0.5]
plt.legend(loc='best', shadow=True )
plt.show()
```
